```python
import pandas as pd
import json
from pprint import pprint
```

```python
with open("majors_to_jobs.json") as json_file:
    m_j = json.load(json_file)

m_j
```

```python
jobs = []

for i in m_j:
    if (i["major_category"] == "Agriculture & Natural Resources"):
        if i['occ_title'] not in jobs:
            jobs.append(i['occ_title'])
```

```python
job_specs = {}

for i in jobs:
#     print(i)
    for j in m_j:
        if (i  == j['occ_title']):
            job_specs[i] = {'p25th':j['p25th'], 'median':j['median'], 'p75th':j['p75th']}
            break
```
