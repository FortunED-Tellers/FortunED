```python
import pandas as pd
import pprint as pp
```

```python
# Read Algo input csv
Algo_input = pd.read_csv("Algo_data_input.csv")
Algo_input
```

```python
target = Algo_input["decision"]
target_names = ["Yes", "No"]
```

```python
data = Algo_input.drop(["decision","time_to_repay"], axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)

# Support vector machine linear classifier
from sklearn.svm import SVC 
model_SVM = SVC(kernel='linear')
model_SVM.fit(X_train, y_train)

# Model Accuracy
# print('Test Acc: %.3f' % model.score(X_test, y_test))

# Calculate classification report
from sklearn.metrics import classification_report
predictions_SVM = model_SVM.predict(X_test)
print("SVM Model Classification Report")
print("------------------------------------------------------")
print(classification_report(y_test, predictions_SVM,
                            target_names=target_names))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5],
              'gamma': [0.0001, 0.001]}
grid = GridSearchCV(model_SVM, param_grid, verbose=3)
grid.fit(X_train, y_train)
predictions_grid = grid.predict(X_test)
print("SVM Model Classification Report with GridSearch")
print(classification_report(y_test, predictions_grid,
                            target_names=target_names))
```

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
predictions_LR = model_LR.predict(X_test)
print("Logistic Regression Model Classification Report")
print("------------------------------------------------------")
print(classification_report(y_test, predictions_LR,
                            target_names=target_names))
```
