```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import re
import json
```

```python
def gather_data(url):
    try:
        base_url = requests.get(url)
    
        soup = BeautifulSoup(base_url.text, 'html.parser')

        income = soup.find_all('tr', class_="odd results")[1]

        count = 0

        for r in income:
            count += 1
            if count == 4:
                wage_t = r.text
        
        wage = re.findall(r'\n(.+?)\n', wage_t)[0][8:]

        state_text = soup.find_all('h1')[1].text

        state = state_text[28:]
    
        main_table = soup.find('table', class_='salary_table').text

        substr = re.findall(r'\n(.+?)\n', main_table)

        mang = substr[2][14:]
        busi = substr[4][14:]
        comp = substr[6][14:]
        arch = substr[8][14:]
        life = substr[10][14:]
        comm = substr[12][14:]
        lega = substr[14][14:]
        educ = substr[16][14:]
        arts = substr[18][14:]
        heat = substr[20][14:]
        heas = substr[22][14:]
        prot = substr[24][14:]
        food = substr[26][14:]
        buil = substr[28][14:]
        pers = substr[30][14:]
        sale = substr[32][14:]
        offi = substr[34][14:]
        farm = substr[36][14:]
        cons = substr[38][14:]
        inst = substr[40][14:]
        prod = substr[42][14:]
        tran = substr[44][14:]

        dic =  {}
        dic1 = {}

        key1 = "living wage"
        value1 = wage
        key2 = "Management"
        value2 = mang
        key3 = "Business & Financial Operations"
        value3 = busi
        key4 = "Computer & Mathematical"
        value4 = comp
        key5 = "Architecture & Engineering"
        value5 = arch
        key6 = "Life, Physical, & Social Science"
        value6 = life
        key7 = "Community & Social Service"
        value7 = comm
        key8 = "Legal"
        value8 = lega
        key9 = "Education, Training, & Library"
        value9 = educ
        key10 = "Arts, Design, Entertainment, Sports, & Media"
        value10 = arts
        key11 = "Healthcare Practitioners & Technical"
        value11 = heat
        key12 = "Healthcare Support"
        value12 = heas
        key13 = "Protective Service"
        value13 = prot
        key14 = "Food Preparation & Serving Related"
        value14 = food
        key15 = "Building & Grounds Cleaning & Maintenance"
        value15 = buil
        key16 = "Personal Care & Service"
        value16 = pers
        key17 = "Sales & Related"
        value17 = sale
        key18 = "Office & Administrative Support"
        value18 = offi
        key19 = "Farming, Fishing, & Forestry"
        value19 = farm
        key20 = "Construction & Extraction"
        value20 = cons
        key21 = "Installation, Maintenance, & Repair"
        value21 = inst
        key22 = "Production"
        value22 = prod
        key23 = "Transportation & Material Moving"
        value23 = tran

        dic[state] = dic1
        dic1[key1] = value1
        dic1[key2] = value2
        dic1[key3] = value3
        dic1[key4] = value4
        dic1[key5] = value5
        dic1[key6] = value6
        dic1[key7] = value7
        dic1[key8] = value8
        dic1[key9] = value9
        dic1[key10] = value10
        dic1[key11] = value11
        dic1[key12] = value12
        dic1[key13] = value13
        dic1[key14] = value14
        dic1[key15] = value15
        dic1[key16] = value16
        dic1[key17] = value17
        dic1[key18] = value18
        dic1[key19] = value19
        dic1[key20] = value20
        dic1[key21] = value21
        dic1[key22] = value22
        dic1[key23] = value23
        
        data.append(dic)
        
        return data
    
    except:
        pass
```

```python
data = []

for i in range(1,57):
    gather_data(f"http://livingwage.mit.edu/states/{i}")
```
