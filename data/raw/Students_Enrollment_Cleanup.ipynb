```python
import pandas as pd
from scipy.stats import linregress
```

```python
students_enrollment_path = 'Students_Enrolement.xlsx'
```

```python
studEnroll_df=pd.read_excel(students_enrollment_path)
studEnroll_df
```

```python
studEnroll_df=studEnroll_df.drop([0,1,2,3,4])
```

```python
studEnroll_df=studEnroll_df.rename(columns=
    {'Table 302.10. Recent high school completers and their enrollment in college, by sex and level of institution: 1960 through 2018':'Year',
     'Unnamed: 1':'HS completers',
     'Unnamed: 7':'Percent of HS completers enrolled in college'
})
studEnroll_df
```

```python
studEnroll_df=studEnroll_df.loc[:,['Year','HS completers','Percent of HS completers enrolled in college']]
```

```python
studEnroll_df=studEnroll_df.dropna()
studEnroll_df
```

```python
studEnroll_df[['Year','Total']]=studEnroll_df['Year'].str.split(' ',expand=True)
```

```python
studEnroll_df['Total']=studEnroll_df['HS completers']*(studEnroll_df['Percent of HS completers enrolled in college']/100)*1000
studEnroll_df['Year']=studEnroll_df['Year'].astype(int)
studEnroll_df['Total']=studEnroll_df['Total'].astype(int)
```

```python
studEnroll_df=studEnroll_df.loc[studEnroll_df['Year']>2006]
studEnroll_df
```

```python
studEnroll_df=studEnroll_df.loc[:,['Year','Total']]
studEnroll_df
```

```python
x_values = studEnroll_df['Year']
y_values = studEnroll_df['Total']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
```

```python
print(f'y={round(slope,2)}x+{round(intercept,1)}')
```

```python
newValues_df = pd.DataFrame({'Year': [2019, 2020]})
```

```python
newValues_df['Total']=slope*newValues_df['Year']+intercept
newValues_df['Total']=newValues_df['Total'].astype(int)
newValues_df
```

```python
EnrolledStud_df=studEnroll_df.append(newValues_df)
EnrolledStud_df=EnrolledStud_df.reset_index()
EnrolledStud_df=EnrolledStud_df.drop(columns=['index'])
```

```python
EnrolledStud_df.to_excel('Enrolled_Students.xlsx',index=False)
```
