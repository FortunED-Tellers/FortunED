```python
import pandas as pd
import json
import pprint as pp
import random
import statistics as st
```

```python
#Read State Living Wage and generate 50 random numbers from livingWage
with open('state_wage_data.json') as json_file: 
        state_wage = json.load(json_file)

Living_Wage = {}        
for data in state_wage.values():
    for values  in data:
        for key,value in values.items():
#             majors = value.keys()
            wage_split = value["living wage"].split("$")[1].split(",")
            wage_join =pd.to_numeric(''.join(map(str, wage_split)),errors='coerce')  
            Living_Wage[key] = wage_join
            
Living_Wage_df = pd.DataFrame(Living_Wage.items())
Living_Wage_df.columns = ["state","living_wage"]
l_min = Living_Wage_df["living_wage"].min()
l_max = Living_Wage_df["living_wage"].max()

living_wage_random = []
for i in range(50):
    num=random.randint(l_min, l_max)
    living_wage_random.append(num)
    
living_wage_random

```

```python
#Read majors and occupation data and generate 50 random numbers from income
with open('majors.json') as json_file: 
        majors_data = json.load(json_file)


income_data = {}

for key,value in majors_data.items():
    for data in value:
        Major_Category = data['Major_Category']
        Possible_occupation = data['Possible_Occupations']
       
        for record in Possible_occupation:
            Occupation = record['Occupation']
            Median_sal_split = record['Median_Occ_Salary'].split(",")
            Median_sal = ''.join(map(str, Median_sal_split))
            Income = pd.to_numeric(Median_sal, errors='coerce')  
            income_data[Occupation] = Income
            data["Occ"] = income_data

            
income_data_df = pd.DataFrame(income_data.items())
income_data_df.columns = ["Occupation","Median_income"]
income_data_df.dropna(inplace=True)


i_min = income_data_df["Median_income"].min()
i_max = income_data_df["Median_income"].max()

income_data_random = []

for i in range(50):
    num=random.randint(i_min, i_max)
    income_data_random.append(num)

income_data_random
```

```python
#Read University Tuition data and generate 50 random numbers from out-of-state Tuition
with open('UniversitiesTuition.json') as json_file: 
        univ_data = json.load(json_file)

univ_Tuition = {}       
for record in univ_data:
    curr_tuition = int(record['Out_2020_2021'])
    university = record['University']
    univ_Tuition[university] = curr_tuition  

univ_data_df = pd.DataFrame(univ_Tuition.items())
univ_data_df.columns = ["Univ","Curr_Tuition"]


u_min = univ_data_df["Curr_Tuition"].min()
u_max = univ_data_df["Curr_Tuition"].max()

debt_data_random = []

for i in range(50):
    num=random.randint(u_min, u_max)
    debt_data_random.append(num)

debt_data_random
```

```python
Algo_data_input = pd.DataFrame({"living_wage":living_wage_random,
                               "income":income_data_random,
                               "debt":debt_data_random})

Algo_data_input["time_to_repay"] = Algo_data_input["debt"]/((Algo_data_input["income"] - Algo_data_input["living_wage"])*0.3)

Algo_data_input['decision'] = ["Yes" if (x>0  and x < 10) else "No" for x in Algo_data_input['time_to_repay']]

Algo_data_input
```

```python
Algo_data_input.to_csv("Algo_data_input.csv",index = False)
```
