```python
import pandas as pd
import json
from pprint import pprint
```

```python
file = pd.read_csv("major_degree_industry_job_all.csv")

file.head()
```

```python
def get_nested_rec(key, grp):
    rec = {}
    rec['major_category'] = grp['major_category']
    rec['major'] = grp['major']
    rec['occ_industry'] = grp['occ_industry']
    rec['occ_title'] = grp['occ_title']
#     rec['degree_required'] = grp['degree_required']
#     rec['work_exp_required'] = grp['work_exp_required']
#     rec['total'] = grp['total']
#     rec['employed'] = grp['employed']
#     rec['emp_full_time_year_round'] = grp['emp_full_time_year_round']
#     rec['unemployed'] = grp['unemployed']
#     rec['unemployment_rate'] = grp['unemployment_rate']
#     rec['p25th'] = grp['p25th']
#     rec['median'] = grp['median']
#     rec['p75th'] = grp['p75th']

    for field in ['major_category','major','occ_industry','occ_title']:
        rec[field] = list(grp[field].unique())

    return rec
```

```python
records = []
for key, grp in file.groupby(['major_category']):
#     print(grp['major'])
    rec = get_nested_rec(key, grp)
    records.append(rec)

records = dict(data = records)
```

```python
pprint(records)
```

```python
with open('majors_only.json', 'w') as f:
    json.dump(records, f)
```
