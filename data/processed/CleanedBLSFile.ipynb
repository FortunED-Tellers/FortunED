```python
# Import dependencies
import pandas as pd
import os
import csv
```

```python
# File to load
bls_file = "occupation_and_wage.csv"

bls_file_df = pd.read_csv(bls_file, encoding="utf-8")
bls_file_df.head()
```

```python
to_drop = ['area', 'area_title', 'area_type', 'naics', 'naics_title', 'i_group', 'own_code', 'o_group', 'h_median', 'h_pct75',
            'h_pct90', 'a_pct10', 'a_pct25', 'a_pct75', 'a_pct90', 'annual', 'hourly']

bls_file_df.drop(to_drop, inplace = True, axis = 1)
bls_file_df.head()
```

```python
to_drop_more = ['tot_emp', 'emp_prse', 'jobs_1000', 'loc_quotient', 'pct_total', 'h_mean', 'a_mean', 'mean_prse',
                'h_pct10', 'h_pct25']
bls_file_df.drop(to_drop_more, inplace = True, axis = 1)
bls_file_df.head()
```

```python
new_names = {'occ_code': 'Occupation Code',
            'occ_title': 'Occupation Title',
            'a_median': 'Annual Median Wage'}
bls_file_df.rename(columns = new_names, inplace = True)
```
